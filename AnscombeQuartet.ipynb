```python
# import dependencies
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Load the example dataset for Anscombe's quartet
df = sns.load_dataset("anscombe")
sns.set(style="ticks")
x  = list(df[0 :11].x)
y1 = list(df[0 :11].y)
y2 = list(df[11:22].y)
y3 = list(df[22:33].y)
x4 = list(df[33:44].x)
y4 = list(df[33:44].y)
```

```python
pd.DataFrame([x,y1,y2,y3,x4,y4],list(['x','y1','y2','y3','x4','y4']))
```

```python
print("Mean of x  is",round(np.mean(x),2))
print("Mean of x4 is",round(np.mean(x4),2))
print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
print("Mean of y1 is",round(np.mean(y1),2))
print("Mean of y2 is",round(np.mean(y2),2))
print("Mean of y3 is",round(np.mean(y3),2))
print("Mean of y4 is",round(np.mean(y4),2))
```

```python
print("Variance of x  is",round(np.var(x),2))
print("Variance of x4 is",round(np.var(x4),2))
print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
print("Variance of y1 is",round(np.var(y1),2))
print("Variance of y2 is",round(np.var(y2),2))
print("Variance of y3 is",round(np.var(y3),2))
print("Variance of y4 is",round(np.var(y4),2))
```

```python
print("Correlation Coefficient between (x1,y1)  is",round(np.corrcoef(x ,y1)[0,1],2))
print("Correlation Coefficient between (x2,y2)  is",round(np.corrcoef(x ,y2)[0,1],2))
print("Correlation Coefficient between (x3,y3)  is",round(np.corrcoef(x ,y3)[0,1],2))
print("Correlation Coefficient between (x4,y4)  is",round(np.corrcoef(x4,y4)[0,1],2))
```
